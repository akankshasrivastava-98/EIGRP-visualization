```python
info_eigrp1 = eigrp1['Info'].to_list()
time1 = eigrp1['Time'].to_list()
```

```python
info_eigrp2 = eigrp2['Info'].to_list()
time2 = eigrp2['Time'].to_list()
```

```python
print('\nInfo Fields of pcap1\n')
for info in info_eigrp1:
    print(info)
#     flag =info.split('')
#     print('\n',flag[0])
```

```python
print('\nInfo Fields of pcap2\n')
for info in info_eigrp2:
    print(info)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(time1, info_eigrp1)
plt.xlabel('Time Interval')
plt.ylabel('Flag')
plt.title('EIGRP Analysis pcap1')
plt.show()
```

```python
plt.plot(time2, info_eigrp2)
plt.xlabel('Time Interval')
plt.ylabel('Flag')
plt.title('EIGRP Analysis pcap2')
plt.show()
```

```python
plt.plot(time1, info_eigrp1,label='pcap1 plot')
plt.plot(time2, info_eigrp2, label='pcap2 plot')
plt.xlabel('Time Interval')
plt.ylabel('Flag')
plt.title('EIGRP Analysis')
plt.legend()
```
